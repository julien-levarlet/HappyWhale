```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("data/common_cropped_train_imgs/0a7a8f558c14a1.jpg")
print(img.shape)
plt.imshow(img)
```

```python
dim = 256,100
resized = cv2.resize(img, dim)
resized.shape
plt.imshow(resized)
```

```python
# test padding
dim = 128,128
out_size = max(resized.shape)
x_pad = (out_size-resized.shape[0])// 2
y_pad = (out_size-resized.shape[1])// 2
pad_size = [[x_pad], [y_pad], [0]]
padded = np.pad(resized, pad_size)
plt.imshow(padded)
```

```python
plt.imshow(cv2.resize(padded, dim))
```
