```python
import pandas as pd
from sklearn.model_selection import train_test_split
df=pd.read_csv("../data/common_train.csv")
print(df.shape)
print(df["individual_id"].nunique())
df.groupby("individual_id").count()
df_train, df_test = train_test_split(df, test_size=0.2, stratify=df["individual_id"])
print(df_train.shape)
print(df_test.shape)
print(df_train.groupby("individual_id").count())
df_test.groupby("individual_id").count()
```
